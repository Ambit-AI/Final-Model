```python
#load dataset from hugging face for fine tuning
from datasets import load_dataset
dataset = load_dataset("mteb/tweet_sentiment_extraction")
```

```python
#display loaded training dataset
dataset
```

```python
#load example of data point in the training dataset
example = dataset['train'][0]
example
```

```python
#load tokenizer, define how the tokenizer work and tokenized a map datasets
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment-latest")

def tokenize_function(examples):
    return tokenizer(examples["text"], padding="max_length", max_length=512, truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
#sectioned the training dataset to speed up fine tuning to 500 datapoint
small_train_dataset = tokenized_datasets["train"].shuffle(seed=42).select(range(500))
small_eval_dataset = tokenized_datasets["test"].shuffle(seed=42).select(range(500))
```

```python
#load model for fine tuning
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained("cardiffnlp/twitter-roberta-base-sentiment-latest", num_labels=3)
```

```python
#load metric
import numpy as np
import evaluate

metric = evaluate.load("accuracy")
```

```python
#how the trainer computer metrics
def compute_metrics(eval_pred):

    logits, labels = eval_pred

    predictions = np.argmax(logits, axis=-1)

    return metric.compute(predictions=predictions, references=labels)
```

```python
#trainign args, save output train model, set evaluation strategy
from transformers import TrainingArguments, Trainer
training_args = TrainingArguments(output_dir='/content/drive/MyDrive/RnD/CheckPoint', evaluation_strategy="epoch")
```

```python
#set up trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics,
)
```

```python
#start training model (this will take about 4h)
trainer.train()
```

GUI STUFF GOES UNDER HERE

```python
import gradio as gr


history = []
total = 0
count = 0
def predict(input_text):
    global history, total, count
    output = fine_tuned(input_text)
    sentiment = output[0]["label"]
    # score = output[0]["score"]
    if sentiment == "positive":
      score = 1
    elif sentiment == "neutral":
      score = 0.5
    else:
      score = 0
    total += score
    count += 1
    history.append(str({input_text:{"sentiment":sentiment, "score":f"{score}"}}))
    average = total/count
    if average > 0.5:
      average_sentiment = "positive"
    elif average == 0.5:
      average_sentiment = "neutral"
    else:
      average_sentiment = "negtive"
    return {"average_sentiment":average_sentiment, "average_score":f"{average:.4f}"}, "\n".join(history)

def clear():
    global history, total, count
    history = []
    count = 0
    total = 0
    return ""

with gr.Blocks(theme=gr.themes.Soft()) as demo:
    input_text = gr.Textbox( label='user', placeholder='input text')
    # model_name = gr.Dropdown(['MultinomialNB', 'LogisticRegression', 'DecisionTreeClassifier', 'RandomForestClassifier', 'SVM'], label='type', value= 1)
    predict_butn = gr.Button("predict")
    output = gr.Textbox(label="output")
    # output2 = gr.DataFrame(label= "probability")
    history_text = gr.Textbox(label="history", placeholder=history)
    clear_butn = gr.Button("clear")
    predict_butn.click(fn=predict, inputs=[input_text], outputs=[output, history_text], api_name="predict")
    clear_butn.click(fn=clear, inputs=[], outputs=[history_text], api_name="clear")

demo.launch(share=True, debug=True)
```
